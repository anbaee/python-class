### Diabet Dataset
#### The variables can be summarized as follows:

**Input Variables** (X):
<ul>
    <li>Number of times pregnant
    <li>Plasma glucose concentration a 2 hours in an oral glucose tolerance test
    <li>Diastolic blood pressure (mm Hg)
    <li>Triceps skin fold thickness (mm)
    <li>Hour serum insulin (mu U/ml)
    <li>Body mass index (weight in kg/(height in m)^2)
    <li>Diabetes pedigree function
    <li>Age (years)
</ul>

**Output Variables** (y):

Class variable (0 or 1)

```python
import pandas as pd
dataset = pd.read_csv('datasetDiabets.csv')
```

```python
print(dataset)
dataset = dataset.dropna()
```

```python
print(dataset.describe())
```

```python
X = dataset.iloc[:,0:8]
y = dataset.iloc[:,8]
```

We can piece it all together by adding each layer:

<ul>
    <li>The model expects rows of data with 8 variables (the input_dim=8 argument)
<li>The first hidden layer has 12 nodes and uses the relu activation function.
<li>The second hidden layer has 8 nodes and uses the relu activation function.
<li>The output layer has one node and uses the sigmoid activation function.
    </ul>

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

<ul>
    <li>We formulate problem as binary classification.
    <li> We can use Binary-crossentropy loss function.
    <li>We can use Adam optimizer based on Gerradiant descent algorithm.
    <li> We can use accuracy as eveluation metric
</ul>

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Training occurs over **epochs** and each epoch is split into **batches**.
<ul>
<li>Epoch: One pass through all of the rows in the training dataset.
<li>Batch: One or more samples considered by the model within an epoch before weights are updated.
    </ul>
    

```python
# fit the keras model on the dataset
history = model.fit(X, y, epochs=150, batch_size=30)
```

```python
accuracy = model.evaluate(X, y)
print(accuracy)
```

```python
predictions = model.predict_classes(X)
# summarize the first 5 cases
for i in range(10):
	print('%s => %d (expected %d)' % (list(X.iloc[i]), predictions[i], y[i]))
```
